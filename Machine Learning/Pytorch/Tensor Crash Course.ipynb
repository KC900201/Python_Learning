```python
# Section 3 V7 - Basic vectorized ops with tensors
t_one = torch.tensor([1, 3, 4])
t_two = torch.tensor([1, 2, 3])
```

```python
print(t_one + t_two)
print(t_one - t_two)
```

```python
#dot product tensors
#multiply elements in tensors col * col and add in total of product
dot_product = torch.dot(t_one, t_two)
# 1*1 + 3*2 + 4*3
print(dot_product)
```

```python
# Secton 3 V10 - Matrix Multiplication
```

```python
mat_a = torch.tensor([0, 3, 45, 5, 5, 2]).view(2, 3)
mat_a
```

```python
mat_b = torch.tensor([3, 4, 4, -2, 4, -2]).view(3, 2)
mat_b
```

```python
mat_c = torch.matmul(mat_a, mat_b) # matrix multiplication
mat_d = mat_a @ mat_b # different operation, same result
print(mat_c)
print(mat_d)
```

```python
# Secton 3 V10 - Gradient with PyTorch (Differentiation)
```

```python
# Example
x = torch.tensor(2.0, requires_grad=True) # 2.0 second derivative, needs gradient
y = 9*x**4 + 2*x**3 + 3*x**2 + 6*x + 1
y.backward() # backward propagation?
#print(y)
x.grad
```

```python
# Partial derivative
x = torch.tensor(1.0, requires_grad=True) 
z = torch.tensor(2.0, requires_grad=True)
y = x**2 + z**3
y.backward()
x.grad
```
